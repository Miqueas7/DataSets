#Ejemplo de Sintaxis

### Crear Sesi√≥n

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('cluster00').getOrCreate()
```

### Cargar Datos

```
ruta = 'dbfs:/FileStore/shared_uploads/jgamarramoreno@gmail.com/sample_kmeans_data.txt'
```

```
datos = spark.read.format('libsvm').load(ruta)
```

```
datos.show()
```

## Crear y Entrenar el modelo KMeans

```
from pyspark.ml.clustering import KMeans
```

```
# Crear el modelo y entrenarlo
kmeans = KMeans().setK(2).setSeed(1)
modelo = kmeans.fit(datos)
```

```
#Evaluar el agrupamiento calculando la Sumatoria de los Errores al cuadrado.
# resumen = modelo.summary
# resumen.trainingCost
```

```
centroides = modelo.clusterCenters()
```

```
print("Centroides de Clusters")
for centroide in centroides:
    print(centroide)
```
