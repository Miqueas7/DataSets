### Creación de un RDD

##### A partir de una lista

```python
# creamos una lista
lista = ['Apache Spark','Analítica de Datos','PySpark','Google Cloud Paltaform']
```

```python
# crear el RDD con parallelize
rdd = spark.sparkContext.parallelize(lista)
```

```python
rdd.collect()
```

##### Crear RDD vacío

```python
rddvacio = spark.sparkContext.emptyRDD()
```

```python
# collect no tiene nada que juntar
rddvacio.collect()
```

#### Crear RDD a partir de un archivo

```python
rdd02 = spark.sparkContext.textFile("gs://analitica_bucket_20220926/datos/persona.data")
```

```python
rdd02.collect()
```

#### Crear RDD a partir de otro RDD con flatMap

```python
rdd03 = rdd02.flatMap(lambda x: x.split("|"))
```

```python
rdd03.collect()
```

#### función map() para agregar columna con el valor 1

```python
rdd04 = rdd03.map(lambda x:(x,1))
```

```python
rdd04.collect()
```

Algunas funciones aplicables a los RDDs:reduceByKey(), sortByKey(), filter(), first(), max(), reduce(), etc.

### DataFrame

Con PySpark se pueden crear DataFrame a partir de fuentes de datos TXT, CSV, JSON, ORV, Avro, Parquet, XML, HDFS, S3, DBFS, Azure Blob, etc.

Para crear un DataFrame a partir de un RDD se usa el método DF()

```python
datos = [("Javier","18"),("Francisco","13"),("Fernando","20")]
```

```python
rdd05 = spark.sparkContext.parallelize(datos)
```

```python
rdd05.collect()
```

```python
# crear el dataframe
df01 = rdd05.toDF()
```

```python
# mostrar el esquema del dataframe
df01.printSchema()
```

```python
# dar nombre a las columnas
columnas = ["Nombre","Nota"]
df01 = rdd05.toDF(columnas)
```

```python
df01.printSchema()
```

##### Método createDataFrame desde una sesión spark

```python
df02 = spark.createDataFrame(datos).toDF(*columnas)
```

```python
df02.show()
```

##### Crear DataFrame con esquema

```python
from pyspark.sql.types import StructType, StructField
```

```python
from pyspark.sql.types import StringType, IntegerType
```

```python
datos02 = [
    ("José","","Cárdenas Rojas","8754","M",5200),
    ("Pedro","Rafael","Gonzáles Prado","9514","M",3500),
    ("Carlos","Gabriel","Solis Castro","8421","M",4000),
    ("Edith","María","Rosales Quintana","9632","F",4000),
    ("Jane","Mary","Rivera Canto","1598","F",-1)
]
```

```python
esquema = StructType([
    StructField("PrimerNombre",StringType(),True),
    StructField("SegundoNombre",StringType(),True),
    StructField("Apellidos",StringType(),True),
    StructField("ID",StringType(),True),
    StructField("Género",StringType(),True),
    StructField("Salario",IntegerType(),True)
])
```

```python
df03 = spark.createDataFrame(data=datos02,schema=esquema)
```

```python
df03.printSchema()
```

```python
df03.show()
```

Creación del dataframe usando csv almacenado en google cloud storage

```python
from pyspark.sql.types import *
```

```python
ruta = "gs://analitica_bucket_20220926/datos/persona.data"
```

```python
esquema_df = StructType([
    StructField("ID",StringType(),True),
    StructField("NOMBRE",StringType(),True),
    StructField("TELEFONO",StringType(),True),
    StructField("CORREO",StringType(),True),
    StructField("FECHA_INGRESO",StringType(),True),
    StructField("EDAD",IntegerType(),True),
    StructField("SALARIO",DoubleType(),True),
    StructField("ID_EMPRESA",StringType(),True)
])
```

```python
df04 = spark.read.format("CSV").option(
    "header","true").option("delimiter","|").schema(esquema_df).load(ruta)
```

```python
df04.show()
```

##### Seleccionar columnas con select()

```python
spark = SparkSession.builder.appName("MiAplicacion").getOrCreate()
```

```python
datos03 = [
    ("José","Sánchez","Perú","Trujillo"),
    ("Carlos","Altamirano","Argentina","Buenos Aires"),
    ("Fernando","Samané","Perú","Trujillo"),
    ("Federico","Chavez","Perú","Lima")
]
columnas = ["nombre","apellido","país","ciudad"]
```

```python
df05 = spark.createDataFrame(data=datos03,schema=columnas)
```

```python
df05.show()
```

Uso de select()

```python
df05.select("nombre","apellido").show()
```

```python
df05.select(df05.nombre,df05.apellido).show()
```

```python
df05.select(df05["nombre"],df05["apellido"]).show()
```

```python
from pyspark.sql.functions import col
df05.select(col("nombre"),col("apellido")).show()
```

##### Función withColumn()

```python
df03.show()
```

```python
datos04 = [
    ("José","","Cárdenas Rojas","1991-04-01","M",5200),
    ("Pedro","Rafael","Gonzáles Prado","2000-05-19","M",3500),
    ("Carlos","Gabriel","Solis Castro","1978-09-05","M",4000),
    ("Edith","María","Rosales Quintana","1967-12-01","F",4000),
    ("Jane","Mary","Rivera Canto","1980-02-17","F",-1)
]
```

```python
columnas = ["prinombre","segnombre","apellido","fec_nac","genero","salario"]
```

```python
df06 = spark.createDataFrame(data=datos04,schema=columnas)
```

```python
df06.show()
```

```python
df06.printSchema()
```

```python
df06.withColumn("salario",col("salario").cast("Integer")).show()
```

```python
df06.printSchema()
```

```python
# Actualizar el valor de salario
df06.withColumn("salario",col("salario")*100).show()
```

```python
df06.show()
```

```python
df06.withColumn("descuento",col("salario")*0.10).show()
```

```python
df06.show()
```

```python
# Renombrar una columna
df06.withColumnRenamed("genero","sexo").show()
```

```python
df06.show()
```

```python
#Eliminar una columna
df06.drop("salario").show()
```

```python
df06.show()
```
