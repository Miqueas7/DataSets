# Ejemplo de Regresión Lineal

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName("ejemplo_rl").getOrCreate()
```

```
from pyspark.ml.regression import LinearRegression
```

```
ruta = "dbfs:/FileStore/shared_uploads/jgamarramoreno@gmail.com/sample_linear_regression_data.txt"
```

```
# cargar los datos
datos = spark.read.format("libsvm").load(ruta)
```

```
datos.show()
```

Este es el formato que Spark espera. Dos columnas con las etiquetas "label" y "features".

#### Modelo de Regresión Lineal

```
modelo_rl = LinearRegression(featuresCol='features',labelCol='label',predictionCol='prediction')
```

```
modelo = modelo_rl.fit(datos)
```

```
modelo.intercept
```

```
modelo.coefficients
```

Resumen del modelo

```
resumen = modelo.summary
```

```
resumen.residuals.show()
```

```
print("RMSE: {}".format(resumen.rootMeanSquaredError))
```

```
print("r2: {}".format(resumen.r2))
```

#### División de datos en conjunto de entrenamiento y prueba

```
datos_recuperados = spark.read.format("libsvm").load(ruta)
```

```
train_data, test_data = datos_recuperados.randomSplit([0.7,0.3],seed=100)
```

```
train_data.show()
```

```
test_data.show()
```

```
datos_caracteristicas = test_data.select('features')
```

```
datos_caracteristicas.show()
```

```
modelo_correcto = modelo_rl.fit(train_data)
```

```
resultados_prueba = modelo_correcto.evaluate(test_data)
```

```
resultados_prueba.residuals.show()
print("RMSE: {}".format(resultados_prueba.rootMeanSquaredError))
```

```
predicciones = modelo_correcto.transform(datos_caracteristicas)
```

```
predicciones.show()
```
